```python
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame
```

```python
dislikes = pd.read_csv('Thesis Survey Data - thesis_dislikes.csv')
no_dislikes = pd.read_csv('Thesis Survey Data - thesis_no_dislikes.csv')
```

```python
no_dislikes.attention_school
```

```python
# CLEAN BOT AND INVALID RESPONSES 

# if overall watch time is less than news watch time 
dislikes = dislikes[dislikes['youtube_time_news'] <= dislikes['youtube_time_overall']]
no_dislikes = no_dislikes[no_dislikes['youtube_time_news'] <= no_dislikes['youtube_time_overall']]

# if attention_school is not Empty or 'I don’t recall how hard I worked'
dislikes = dislikes[(dislikes['attention_school'] == 'I don’t recall how hard I worked') | (dislikes['attention_school'].isnull())]
no_dislikes = no_dislikes[(no_dislikes['attention_school'] == 'I don’t recall how hard I worked') | (no_dislikes['attention_school'].isnull())]

# if attention_carat is not Empty or 'tarac'
dislikes = dislikes[(dislikes['attention_carat'] == 'tarac') | (dislikes['attention_carat'].isnull())]
no_dislikes = no_dislikes[(no_dislikes['attention_carat'] == 'tarac') | (no_dislikes['attention_carat'].isnull())]

# if duration is less than 100s or more than 1200s
dislikes = dislikes[(dislikes['duration'] >= 100) | (dislikes['duration'] <= 1200)]
no_dislikes = no_dislikes[(no_dislikes['duration'] >= 100) | (no_dislikes['duration'] <= 1200)]
```

```python
# PLOT LOCATION DISTRIBUTION

geometry = [Point(xy) for xy in zip(dislikes['longitude'], dislikes['latitude'])]
gdf_dislikes = GeoDataFrame(dislikes, geometry=geometry) 

geometry = [Point(xy) for xy in zip(no_dislikes['longitude'], no_dislikes['latitude'])]
gdf_no_dislikes = GeoDataFrame(no_dislikes, geometry=geometry)   
```
